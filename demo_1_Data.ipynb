## data_vfseries(): Combined Visual Field Series for General Progression Method (from vfprogression package)

### Description: A data frame sample including the following visual field related measurement for two eyes, each with 10 follow-ups. The columns are described in follows:
1. eyeid: eyeid, labeled as 1,2... for different eyes.
2. nvisit: number of visits.
3. yearsfollowed: follow-up years.
4. distprevn: to be updated.
5. age: in years.
6. righteye: 1 as right eye, 0 as left eye.
7. malfixrate: VF test malfixation rate.
8. ght: glaucoma hemifield test result.
9. vfi: visual field index.
10. md: mean deviation.
11. mdprob: mean deviation probability.
12. psd: pattern standard deviation.
13. psdprob: pattern standard deviation probability.
14. s1-s54: 52 sensitivity measurements.
15. td1-td54: 52 total deviation measurements.
16. tdp1-tdp54: 52 total deviation probability measurements.
17. pdp1-pdp54: 52 pattern deviation probability measurements. ...

```python
from PyVisualFields import vfprogression
import pandas 

df_VFs_py = vfprogression.data_vfseries()
df_VFs_py.head()
```
